```python
# A class common to all neurons hence representing a general neuron. 
class Neuron:
    def __init__(self, firing_rate=0.0): # We assume that if no input was given to the neuron - the firing rate would be zero.
        self.firing_rate = firing_rate

    # Calculates the firing rate of the neuron based on its stimulus. 
    def Activate(self, stimulus_strength=0.0): # Default stimulus level is zero if not mentioned otherwise.
        self.firing_rate = stimulus_strength # No specific calculation was given in the assignment so I assumed the firing rate is equal to its stimulus. 
        return f"The firing rate of the neuron is {self.firing_rate}"

# Sensory neuron (inherits from the class Neuron)
class SensoryNeuron(Neuron):
    def __init__(self, firing_rate=0.0, receptor_type="No Receptor"): # If no receptor is mentioned: the default text would be 'No Receptor'. 
        super().__init__(firing_rate) # Inherits firing_rate from the previous class. 
        self.receptor_type = receptor_type

    def SenseStimulus(self, stimulus=0.0):
        return f"Receptor type: {self.receptor_type}. Stimulus: {stimulus}"

# Motor neuron (inherits from the class Neuron)
class MotorNeuron(Neuron):
    def __init__(self, firing_rate=0.0, target_muscle="No Muscle"):  # If no receptor is mentioned: the default text would be 'No Muscle'.
        super().__init__(firing_rate) # Inherits firing_rate from the previous class. 
        self.target_muscle = target_muscle

    def ControlMuscle(self):
        return f"Controlling muscle {self.target_muscle} at a {self.firing_rate} firing rate"

# Sensory neuron sensetive solely to light (inherits from the class SensoryNeuron).
class Photoreceptor(SensoryNeuron):
    def __init__(self, firing_rate=0.0):
        super().__init__(firing_rate, receptor_type="light")

    def light_detector(self, light_intensity=0.0):
        self.firing_rate = light_intensity # No specific calculation was given in the assignment so I assumed the firing rate is equal to its stimulus. 
        return f"The firing rate of the photoreceptor is {self.firing_rate}"

# Sensory neuron sensetive solely to mechanical pressure (inherits from the class SensoryNeuron).
class Mechanoreceptor(SensoryNeuron):
    def __init__(self, firing_rate=0.0):
        super().__init__(firing_rate, receptor_type="pressure") 

    def pressure_detector(self, pressure_level=0.0):
        self.firing_rate = pressure_level # No specific calculation was given in the assignment so I assumed the firing rate is equal to its stimulus. 
        return f"The firing rate of the mechanoreceptor is {self.firing_rate}"

# Motor neuron type Alpha (inherits from the class MotorNeuron).
class AlphaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate=0.0):
        super().__init__(firing_rate, target_muscle="skeletal muscle") 

    def skeletal_muscle_control(self):
        return f"Alpha motor neuron activating skeletal muscle with level: {self.target_muscle}"

# Motor neuron type Gamma (inherits from the class MotorNeuron).
class GammaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate=0.0):
        super().__init__(firing_rate, target_muscle="muscle spindle") 

    def muscle_spindle_control(self):
        return f"Gamma motor neuron controlling muscle spindle with level: {self.target_muscle}"
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Settling initial parameters: 
time_duration = 1 # Sec
frequency = 5 # Hz
sample_rate =  100 # How many samples are in one second
modifier = 20 # Percetnage of downregulation 


# Time array
sample_rate = 1/sample_rate
samples = np.arange(0, time_duration + sample_rate, sample_rate)


# The sin data ready for plotting.
sin_plot = np.sin(2 * np.pi * frequency * samples) # We use 2 becuase they correspond to one full sinusoidal circle. 


plt.figure(figsize=(16, 5)) # Figuring the size of the plots. 

# Plotting the original sin graph.
plt.subplot(1, 2, 1) 
plt.plot(samples, sin_plot, color="blue")
plt.title("Original Sine Wave")
plt.xlabel("Time (Sec)")
plt.ylabel("Amplitude")
plt.grid(True)


# Downsample the wave to 20 percent
modifier = int(100/modifier)
modified_samples = samples[::modifier]
modified_sin_plot = sin_plot[::modifier]


# Plotting the downregulated sin graph.
plt.subplot(1, 2, 2) 
plt.plot(modified_samples, modified_sin_plot, color="red")
plt.title("Downsampled Sine Wave")
plt.xlabel("Time (Sec)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()


# Verifying whether the maximum and minimum points are the same even after downregulation. 
if max(modified_sin_plot) == max(sin_plot) and min(modified_sin_plot) == min(sin_plot):
    print("Maxima and Minima are indeed maintained after downregulation.")
else:
    print("something is wrong, Maxima and Minima are not the same after downregulation.")

print("Original min:", min(sin_plot), "| Downsampled min:", min(modified_sin_plot))
print("Original max:", max(sin_plot), "| Downsampled max:", max(modified_sin_plot))
```

### Extracting the data from the excel file

```python
import pandas as pd
import os

# The directory of the file in which our code is placed. 
file_directory = os.getcwd()

# The directory of the excel we were using.
excel_directory = file_directory + '\laptop_price - dataset.csv'  

# Data of the excel file. 
data = pd.read_csv(excel_directory)

print(data)
```

### Plotting the price of all the laptops

```python
# Grouping the values under the column of the company's names that are of the same brand and calculates the average price for all of its laptops. 
average_company_price = data.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False) # Sorting the values in a non-ascending order. 

# Plotting the average proce for each company on the figure. 
average_company_price.plot(kind='bar', edgecolor='black')  

# Settling titles and showing the graph.
plt.title('Average price of laptop for each company')
plt.xlabel('Name of Company')  
plt.ylabel('Average Price (in Euros)')  
plt.show()  
```

### Which company has on average the most expensive laptop? What is the average laptop price for each company?

### Finding the different types of Operating systems presented in the data

```python
# List of all the Operating systems in the excel (including the different type of each OS).
os_list = data['OpSys'].unique()

# Define a mapping of OS variants to their groups.
os_mapping = {
    'Windows 10': 'Windows',
    'Windows 10 S': 'Windows',
    'Windows 7': 'Windows',
    'Mac OS X': 'macOS',
    'macOS': 'macOS',
    'Linux': 'Linux',
    'Android': 'Android',
    'Chrome OS': 'Chrome OS',
    'No OS': 'No OS'
}

# creates a new list by replacing each item in os_list with its mapped value from os_mapping, then removes duplicates using set().m
new_os_list = list(set([os_mapping[i] for i in os_list])) # Using list() makes it more convenient to work with. 

print('The types of Operating Systems are:', ', '.join(new_os_list) + '.') # join() is used to print all the objects from the list in one sentence. 
```

### Plotting for each of the Operating System types the distribution of the prices

```python
#Create a figure with a suitable layout.
plt.figure(figsize = (20, 15))

# Running the for function on the list new_os_list, so num_plots is the number of the index in the list and os_type is the name of the object in it.
for num_plots, os_type in enumerate(new_os_list):
    
    # Determines the number of plots and where they'd be depicted.
    plt.subplot(3, 2, num_plots+1)  # Adding one to num_plots was necessary because it starts as 0 and its not suitable for subplot().  

    # os_prices contains an array of all the prices for all the laptops from the same type of OS (os_type).
    os_prices = data[data["OpSys"].map(os_mapping) == os_type]["Price (Euro)"] # os_mapping was defined in the previous section. 
    
    # Settling the names of the titles and some aesthetics.
    plt.hist(os_prices, edgecolor="black")
    plt.title(f"{os_type} Price Distribution")
    plt.xlabel("Price (Euro)")
    plt.ylabel("Frequency")
    
plt.suptitle("Distribution prices for each Operating System", fontsize=30)
plt.show()
```

```python
import seaborn as sns

# A list of all RAM. 
RAM_data = list(data['RAM (GB)'])

# A list of all the prices. 
price_data = list(data['Price (Euro)'])

# Calculate Pearson's correlation coefficient.
pearson_r = np.corrcoef(RAM_data, price_data)[0,1]

# Determining whether pearson's r is negative or positive. 
if pearson_r > 0:
    x = 'positive'
else:
    x = 'negative' 

# Using it just for the sake of convenience because this sentence is repetitive bellow.
sentence = 'correlation between RAM and computer price.'   


# Printing how strong the correlation was between RAM and computer price and whether it was positive or negative correlation. 
r = abs(pearson_r) # We use the absolute value of r so we won't have to right the whole algorithm twice for both situations (where it is negative and positive).
if r == 0:
    print('No', sentence ) 
elif 0<r<=0.1:
    print('Very Weak', x, sentence)
elif 0.1<r<=0.3:
    print('Weak', x, sentence)
elif 0.3<r<=0.5:
    print('Moderate', x, sentence)
elif 0.5<r<=0.7:
    print('Strong', x, sentence)
elif 0.7<r<1:
    print('Very strong', x, sentence)
elif r == 1:
    print ('Perfect', sentence)


print("Pearson's r:", pearson_r)


# Scatter plot with regression line using Seaborn.
sns.regplot(x=RAM_data, y=price_data, scatter_kws={'color': 'blue', 's': 50}, line_kws={'color': 'red'})  

# Plot customization
plt.title('Price by RAM Size')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (in Euros)')
plt.show()
```

## Question 1: What are the distributions of the prices for each company by each RAM size?

```python

# Group by 'Company' and 'RAM (GB)', and includes the average price for each group.
ram_price_average = data.groupby(['Company', 'RAM (GB)'])['Price (Euro)'].mean().sort_values(ascending=False).reset_index()


ram_sizes = sorted(data['RAM (GB)'].unique())

# Calculate number of rows needed (3 plots per row)
ncols = 2
nrows = (len(ram_sizes) + ncols - 1) // ncols  # This rounds up to ensure all plots fit.

# Loop through each RAM size to create a separate plot.
for RAM in sorted(data['RAM (GB)'].unique()): 

    # Filtering the data so it would include the price of each company at a specific RAM size.
    company_price = ram_price_average[ram_price_average['RAM (GB)'] == RAM][['Company', 'Price (Euro)']]

    # Plot the results.
    company_price.plot(x='Company', y='Price (Euro)', kind='bar', edgecolor='black', legend=False)
    plt.title(f'Average Laptop Price by Company for {RAM} GB RAM')
    plt.xlabel('Laptop Company')
    plt.ylabel("Average Price (Euro)")

plt.show()
```
